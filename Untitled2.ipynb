```python
# ==========================================
# LAST MILE DELIVERY ANALYSIS DASHBOARD
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
from google.colab import files

# ==========================================
# PROJECT HEADER
# ==========================================

print("="*60)
print("LAST MILE DELIVERY ANALYSIS DASHBOARD")
print("="*60)

# ==========================================
# UPLOAD DATASET
# ==========================================

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

print("\nDataset Loaded Successfully")
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nFirst 5 Records")
print(df.head())

# ==========================================
# DATA CLEANING
# ==========================================

df.drop_duplicates(inplace=True)

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# Convert time column
df["hour"] = pd.to_datetime(
    df["order_time"],
    format="%H:%M"
).dt.hour

print("\nData Cleaning Completed")

# ==========================================
# DELAY CATEGORY ANALYSIS
# ==========================================

df["delay_category"] = pd.cut(
    df["delay_mins"],
    bins=[0,10,20,100],
    labels=[
        "Low Delay",
        "Medium Delay",
        "High Delay"
    ]
)

print("\nDelay Category Distribution")
print(df["delay_category"].value_counts())

# ==========================================
# Q1 PEAK HOUR DELAY PATTERN
# ==========================================

print("\n" + "="*60)
print("Q1 PEAK HOUR DELAY ANALYSIS")
print("="*60)

def peak_hour(hour):
    if (8 <= hour <= 10) or (17 <= hour <= 20):
        return "Peak"
    return "Off-Peak"

df["time_category"] = df["hour"].apply(peak_hour)

peak = df[df["time_category"]=="Peak"]["delay_mins"]
offpeak = df[df["time_category"]=="Off-Peak"]["delay_mins"]

print("Peak Average Delay:",
      round(peak.mean(),2))

print("Off Peak Average Delay:",
      round(offpeak.mean(),2))

print("Difference:",
      round(peak.mean()-offpeak.mean(),2))

t,p = ttest_ind(peak,offpeak)

print("P Value:",p)

if p < 0.05:
    print("Significant Difference Found")
else:
    print("No Significant Difference")

plt.figure(figsize=(8,5))
sns.boxplot(
    x="time_category",
    y="delay_mins",
    data=df
)
plt.title("Peak vs Off Peak Delay")
plt.show()

# ==========================================
# Q2 WEATHER VS DELAY
# ==========================================

print("\n" + "="*60)
print("Q2 WEATHER IMPACT ANALYSIS")
print("="*60)

weather_delay = (
    df.groupby("weather_condition")
    ["delay_mins"]
    .median()
)

print(weather_delay)

plt.figure(figsize=(8,5))
sns.boxplot(
    x="weather_condition",
    y="delay_mins",
    data=df
)
plt.title("Weather vs Delay")
plt.show()

rain_df = df[
    df["weather_condition"]=="Rain"
]

rain_impact = (
    rain_df.groupby("order_type")
    ["delay_mins"]
    .median()
)

print("\nRain Impact by Order Type")
print(rain_impact.sort_values(
    ascending=False
))

plt.figure(figsize=(8,5))
rain_impact.sort_values().plot(
    kind="bar"
)
plt.title(
    "Order Type Impact During Rain"
)
plt.ylabel("Median Delay")
plt.show()

# ==========================================
# Q3 RIDER EXPERIENCE EFFECT
# ==========================================

print("\n" + "="*60)
print("Q3 RIDER EXPERIENCE ANALYSIS")
print("="*60)

less_exp = df[
    df["rider_experience_yrs"] < 2
]["delay_mins"]

more_exp = df[
    df["rider_experience_yrs"] > 4
]["delay_mins"]

print(
    "Average Delay (<2 years):",
    round(less_exp.mean(),2)
)

print(
    "Average Delay (>4 years):",
    round(more_exp.mean(),2)
)

t,p = ttest_ind(
    less_exp,
    more_exp
)

print("P Value:",p)

if p < 0.05:
    print(
        "Experience Significantly Affects Delay"
    )
else:
    print(
        "No Significant Difference"
    )

# ==========================================
# CORRELATION ANALYSIS
# ==========================================

print("\n" + "="*60)
print("CORRELATION ANALYSIS")
print("="*60)

numeric_cols = df.select_dtypes(
    include=np.number
)

corr = numeric_cols.corr()

plt.figure(figsize=(8,6))
sns.heatmap(
    corr,
    annot=True
)
plt.title("Correlation Matrix")
plt.show()

# ==========================================
# Q4 DASHBOARD METRICS
# ==========================================

print("\n" + "="*60)
print("Q4 DASHBOARD METRICS")
print("="*60)

df["on_time"] = np.where(
    df["delay_mins"] <= 10,
    1,
    0
)

city_rate = (
    df.groupby("city")
    ["on_time"]
    .mean()*100
)

vehicle_delay = (
    df.groupby("vehicle_type")
    ["delay_mins"]
    .mean()
)

print("\nCity Wise On Time Rate")
print(city_rate)

best_city = city_rate.idxmax()
worst_city = city_rate.idxmin()

print("\nBest City:", best_city)
print("Worst City:", worst_city)

best_vehicle = vehicle_delay.idxmin()

print("Best Vehicle:", best_vehicle)

# ==========================================
# CITY PERFORMANCE
# ==========================================

plt.figure(figsize=(10,5))
city_rate.sort_values().plot(
    kind="bar"
)
plt.title(
    "City Wise On-Time Rate"
)
plt.ylabel("Percentage")
plt.show()

# ==========================================
# VEHICLE COMPARISON
# ==========================================

plt.figure(figsize=(10,5))
vehicle_delay.sort_values().plot(
    kind="bar"
)
plt.title(
    "Vehicle Type Comparison"
)
plt.ylabel("Average Delay")
plt.show()

# ==========================================
# MONTHLY TREND
# ==========================================

df["order_date"] = pd.to_datetime(
    df["order_date"]
)

monthly = (
    df.groupby(
        df["order_date"].dt.month
    )["delay_mins"]
    .mean()
)

plt.figure(figsize=(10,5))
monthly.plot(marker="o")
plt.title(
    "Monthly Delay Trend"
)
plt.xlabel("Month")
plt.ylabel("Average Delay")
plt.grid(True)
plt.show()

# ==========================================
# EXPORT CLEANED DATA
# ==========================================

df.to_csv(
    "cleaned_last_mile_delivery.csv",
    index=False
)

print("\nCleaned Dataset Saved")

print("\nPROJECT COMPLETED SUCCESSFULLY")
```
